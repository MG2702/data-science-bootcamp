```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

```python
plt.plot([0,1], [0,3])
```

```python
plt.plot([0,1], [0,3], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('wykres')
plt.legend()
```

```python
plt.plot([0,1, 2], [0,3,2], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('wykres')
plt.legend()
```

```python
x=np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin, label='x_sin')
plt.plot(x, x_cos, label='x_cos')
plt.legend()
```

```python
plt.subplot(121)
plt.plot([0, 3, 5], [2, 7, 5], 'b--')

plt.subplot(122)
plt.plot([1, 3, 4], [5, 1, 6], 'r--')
```

```python
fig = plt.figure(figsize=(12,8))

plt.subplot(131)
plt.plot([0, 3, 5], [2, 7, 5], 'y-')

plt.subplot(132)
plt.plot([1, 3, 4], [5, 1, 6], 'r--')

plt.subplot(133)
plt.plot([1,3,4], [5,1,6], 'go')
```

```python
fig = plt.figure(figsize=(12,10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
  plt.subplot(220 + idx + 1)
  plt.imshow(plt.imread(img))
```
